# Importing Libarary and Dataset

```python
import pandas as pd
data=pd.read_csv("banknifty.csv")
```

```python
data=data[['Date/Time','Open','High','Low','Close']]
```

# Calculate Perivious Close(Cp)

```python
data['Cp'] = data['Close'].shift(1)
```

```python
data.head(5)
```

# Calculate TR

```python
def Tr(H,L,Cp):
    return max((H-L),abs(H-Cp),abs(L-Cp))
```

```python
length = len(data)
func_outcome = []
for i in range(length):
    outcome = Tr(data.loc[i,'High'],data.loc[i,'Low'],data.loc[i,'Cp'])
    func_outcome.append(outcome)  
data['TR'] = pd.DataFrame(func_outcome)    
```

```python
data.head(5)
```

# Calulate ATR

```python
data['ATR'] = data['TR'].rolling(window=5).mean()
data.head(10)
```

# Claculate Basic Upper & Lower Band

```python
# Finding Upper Band
def UpperB(H,L,M,Atr):
    return (((H + L) / 2)+ (M*Atr))

# Finding Lower Band
def LowerB(H,L,M,Atr):
    return (((H + L) / 2) - (M*Atr))
```

```python
upper_band = []
lower_band = []
M = 1 
M = 1
for i in range(length):
    upper = UpperB(data.loc[i,'High'],data.loc[i,'Low'],M,data.loc[i,'ATR'])
    lower = LowerB(data.loc[i,'High'],data.loc[i,'Low'],M,data.loc[i,'ATR'])
    upper_band.append(upper)
    lower_band.append(lower)
    
data['Basic_Upper_Band'] = pd.DataFrame(upper_band)
data['Basic_Lower_Band'] = pd.DataFrame(lower_band)
```

```python
data.head(10)
```

# Filling The NAN Values

```python
data.fillna(0,inplace=True)
```

# Calculate Final Upper & Lower Band

```python
for i, row in data.iterrows():
    if i==0:
        data.loc[i,"Final_Upper_Band"]=0.00
    else:
        if (data.loc[i,"Basic_Upper_Band"]<data.loc[i-1,"Final_Upper_Band"])|(data.loc[i-1,"Close"]>data.loc[i-1,"Final_Upper_Band"]):
            data.loc[i,"Final_Upper_Band"]=data.loc[i,"Basic_Upper_Band"]
        else:
            data.loc[i,"Final_Upper_Band"]=data.loc[i-1,"Final_Upper_Band"]
```

```python
for i, row in data.iterrows():
    if i==0:
        data.loc[i,"Final_Lower_Band"]=0.00
    else:
        if (data.loc[i,"Basic_Lower_Band"]>data.loc[i-1,"Final_Lower_Band"])|(data.loc[i-1,"Close"]<data.loc[i-1,"Final_Lower_Band"]):
            data.loc[i,"Final_Lower_Band"]=data.loc[i,"Basic_Lower_Band"]
        else:
            data.loc[i,"Final_Lower_Band"]=data.loc[i-1,"Final_Lower_Band"]
```

```python
data.head(19)
```

# Calculate Super Trend Values

```python
for i, row in data.iterrows():
    if i==0:
        data.loc[i,"Super_Trend"]=0.00
    elif (data.loc[i-1,"Super_Trend"]==data.loc[i-1,"Final_Upper_Band"]) & (data.loc[i,"Close"]<=data.loc[i,"Final_Upper_Band"]):
        data.loc[i,"Super_Trend"]=data.loc[i,"Final_Upper_Band"]
    elif (data.loc[i-1,"Super_Trend"]==data.loc[i-1,"Final_Upper_Band"])&(data.loc[i,"Close"]>data.loc[i,"Final_Upper_Band"]):
        data.loc[i,"Super_Trend"]=data.loc[i,"Final_Lower_Band"]
    elif (data.loc[i-1,"Super_Trend"]==data.loc[i-1,"Final_Lower_Band"])&(data.loc[i,"Close"]>=data.loc[i,"Final_Lower_Band"]):
        data.loc[i,"Super_Trend"]=data.loc[i,"Final_Lower_Band"]
    elif (data.loc[i-1,"Super_Trend"]==data.loc[i-1,"Final_Lower_Band"])&(data.loc[i,"Close"]<data.loc[i,"Final_Lower_Band"]):
        data.loc[i,"Super_Trend"]=data.loc[i,"Final_Upper_Band"]
```

# Find Out Market Trend

```python
for i in range(len(data)):
    if data.loc[i,"Close"] > data.loc[i,"Super_Trend"]:
        print(data.loc[i,"Date/Time"],"Buy")
    else:
        print(data.loc[i,"Date/Time"],'Sell')
```

```python
for i, row in data.iterrows():
    if i==0:
        data["ST_BUY_SELL"]="NA"
    elif (data.loc[i,"Super_Trend"]<data.loc[i,"Close"]) :
        data.loc[i,"ST_BUY_SELL"]="BUY"
    else:
        data.loc[i,"ST_BUY_SELL"]="SELL"
```
